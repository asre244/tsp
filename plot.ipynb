```python
import matplotlib.pyplot as plt

with open(
        r'C:\Users\AjithSreenivasan\OneDrive - Robinson Bowmaker Paul\Coursera\Discrete Optimization\tsp\data\tsp_100_3',
        'r') as input_data_file:
    input_data = input_data_file.read()
data_list = input_data.split()
data_list = list(map(float, data_list))
n = int(data_list[0])
points = [i for i in range(0, n)]
coordinates = data_list[1:]
coordinate_list = [(coordinates[i * 2], coordinates[(i * 2) + 1]) for i in range(0, n)]
consolidated_list = [(points[i], coordinate_list[i]) for i in range(0, n)]

# Separate the list of coordinates into two lists: x and y
x_coords, y_coords = zip(*coordinate_list)
```

```python
result = [48, 32, 0, 33, 5, 2, 28, 45, 9, 10, 3, 46, 24, 34, 23, 12, 30, 36, 6, 26, 47, 27, 41, 8, 4, 35, 13, 7, 19, 40, 18, 11, 42, 29, 43, 21, 37, 20, 25, 1, 22, 31, 39, 50, 38, 15, 44, 16, 14, 49, 17]
```

```python
# Create the plot
plt.figure(figsize=(8, 6))
plt.scatter(x_coords, y_coords, c='red', marker='o')

# Add titles and labels
plt.title('Scatter Plot of Coordinates')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Annotate points with their designations
for point, x, y in zip(points, x_coords, y_coords):
    plt.annotate(point, (x, y), textcoords="offset points", xytext=(5,5), ha='center')

# Show the plot
plt.grid(True)
plt.show()
```

```python
# Create the plot
plt.figure(figsize=(8, 6))
plt.scatter(x_coords, y_coords, c='red', marker='o')

# Add titles and labels
plt.title('Scatter Plot of Coordinates')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Annotate points with their designations
for point, x, y in zip(points, x_coords, y_coords):
    plt.annotate(point, (x, y), textcoords="offset points", xytext=(5,5), ha='center')

# Draw lines with arrows according to the order in result
for i in range(0, len(result) - 1):
    start = result[i]
    end = result[i + 1]
    plt.annotate('', xy=(x_coords[end], y_coords[end]), xytext=(x_coords[start], y_coords[start]),
                 arrowprops=dict(facecolor='black', shrink=0.05))

# Show the plot
plt.grid(True)
plt.show()
```
